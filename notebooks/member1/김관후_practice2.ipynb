# H&M 기업이란?
H&M은 스웨덴에서 시작한 글로벌 패션 리테일 기업으로, 트렌드를 빠르게 반영한 상품을 비교적 합리적인 가격대로 제공하는 대형 SPA 브랜드다. 전 세계 오프라인 매장과 온라인 채널을 동시에 운영하며, 시즌·트렌드 변화에 맞춰 상품을 기획하고 빠르게 공급하는 방식으로 시장 경쟁력을 유지해왔다. 최근에는 단순히 저렴하고 빠른 패션을 넘어, 디지털 전환과 지속가능성을 핵심 축으로 기업 체질을 강화하는 방향으로 움직이고 있다.

# 2019년 H&M이 강조한 전략은 크게 3가지!!!
1) 영업이익 개선: 재고 관리 & 마크다운(할인) 감소 효과: 재고를 더 효율적으로 운영하고 불필요한 할인(마크다운)을 줄여 수익성(영업이익)을 개선하려는 방향→ 많이 파는 것뿐 아니라 수익성까지 챙기는 운영 전략이 핵심이다!!!
2) 디지털 전환 & 고객 경험 개선온라인 쇼핑 경험 강화, 모바일 앱 기반 경험 개선→ 온라인 성장(+24%)과도 연결됨! 고객 접점이 매장 중심에서 모바일/온라인 중심으로 이동하는 흐름에 대응한 전략이라고 볼 수 있다!
3) 지속가능성 & 환경 전략: 순환 패션(순환 모델) 강화단순 생산·판매 중심이 아니라 환경/지속가능 관점에서 사업 모델을 강화→ H&M이 장기적으로 브랜드 신뢰 + 소비자 기대에 대응하기 위한 핵심!!!

1) 매출 증가 대비 영업마진 하락 가능성영업이익은 늘었어도 마진율은 과거 수준 대비 낮아지고 하락 추세 가능
→ 성장과 수익성 유지가 동시에 어려운 구조임

2) 경쟁 심화 & 소비자 기대 변화온라인 경쟁 심화, 소비자의 지속가능성/윤리적 소비 요구 증가

3) 소비 패턴 변동성계절성, 유행 변화가 빠른 패션 산업 특성상 수요 예측 실패 시 재고/할인 리스크 커진다

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 0) 타입/기본 정리

ts = ts.copy()
cs = cs.copy()
ac = ac.copy()

# Revenue
ts["Revenue"] = pd.to_numeric(ts["Revenue"], errors="coerce").fillna(0)

# 채널 라벨 (H&M Kaggle 기준: 1=store(오프라인), 2=online(온라인))
channel_map = {1: "오프라인(매장)", 2: "온라인"}
ts["Channel"] = ts["Sales_Channel_Id"].map(channel_map).fillna("기타")

# 월 컬럼 정리
if "Year_Month" in ts.columns:
    ts["Year_Month_dt"] = pd.to_datetime(ts["Year_Month"] + "-01", errors="coerce")
elif "t_dat" in ts.columns:
    ts["Year_Month_dt"] = pd.to_datetime(ts["t_dat"], errors="coerce").dt.to_period("M").dt.to_timestamp()
else:
    ts["Year_Month_dt"] = pd.NaT

# 1) 필요한 컬럼만 추려서 조인 
cs_cols = [c for c in ["Customer_Id","Age_Band","Age","Fn_Label","Active_Label","Club_Member_Status","Fashion_News_Frequency"] if c in cs.columns]
ac_cols = [c for c in ["Article_Id","Product_Group_Name","Garment_Group_Info","Section_Info","Department_Info","Index_Info"] if c in ac.columns]

df = ts.merge(cs[cs_cols], on="Customer_Id", how="left")
df = df.merge(ac[ac_cols], on="Article_Id", how="left")

# 상품군 대표 컬럼 하나로 통일
product_group_col = None
for cand in ["Garment_Group_Info", "Product_Group_Name", "Section_Info", "Department_Info", "Index_Info"]:
    if cand in df.columns:
        product_group_col = cand
        break

df["Product_Group"] = df[product_group_col].fillna("Unknown") if product_group_col else "Unknown"

print("df shape:", df.shape)
print("Product_Group 기준 컬럼:", product_group_col)
print("Channel unique:", df["Channel"].value_counts(dropna=False))
```

```python
# 판매량(아이템수)
p_cnt = df_w.pivot_table(index="month", columns="Channel", values="Customer_Id", aggfunc="size", fill_value=0)\
           .reindex(order_month)
display(p_cnt)

p_cnt.rename(index=month_label).plot(kind="bar")
plt.title("겨울 월별 판매량: 온/오프라인 비교(아이템 기준)")
plt.xlabel("월")
plt.ylabel("판매량(건수)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 채널별 핵심 분해 KPI
base = df_w.groupby("Channel").agg(
    판매량=("Customer_Id", "size"),
    구매고객수=("Customer_Id", "nunique")
)
base["고객당구매수(아이템)"] = base["판매량"] / base["구매고객수"]

# 재구매율(근사): 채널 내 2회 이상 구매 고객 비중
cust_cnt = df_w.groupby(["Channel","Customer_Id"]).size().reset_index(name="cnt")
rep = cust_cnt.assign(repeat=lambda x: x["cnt"]>=2).groupby("Channel")["repeat"].mean()
base["재구매율(근사)"] = rep

# 매출/단가
if "Revenue" in df_w.columns and df_w["Revenue"].notna().any():
    rev = df_w.groupby("Channel").agg(매출=("Revenue","sum"))
    base = base.join(rev)
    base["고객당매출"] = base["매출"] / base["구매고객수"]
    base["평균단가(아이템당)"] = base["매출"] / base["판매량"]

display(base)

# 시각화(온/오프라인 비교가 보이게)
base[["구매고객수","고객당구매수(아이템)","재구매율(근사)"]].plot(kind="bar")
plt.title("겨울 채널 KPI 분해(온/오프라인 비교)")
plt.xlabel("채널")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

if "매출" in base.columns:
    base[["매출","고객당매출","평균단가(아이템당)"]].plot(kind="bar")
    plt.title("겨울 채널 매출 분해(온/오프라인 비교)")
    plt.xlabel("채널")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
TOP_N = 8

# 채널별 Top 상품군 뽑기
top_pg_by_ch = (df_w.groupby(["Channel","Product_Group"]).size()
                .reset_index(name="cnt")
                .sort_values(["Channel","cnt"], ascending=[True, False]))

top_list = top_pg_by_ch.groupby("Channel").head(TOP_N)
display(top_list)

# 월×채널×(Top 상품군) 판매량 피벗
top_groups = top_list["Product_Group"].unique().tolist()
df_w_top = df_w[df_w["Product_Group"].isin(top_groups)].copy()

p_pg = df_w_top.pivot_table(index=["month"], columns=["Channel","Product_Group"],
                            values="Customer_Id", aggfunc="size", fill_value=0).reindex(order_month)

display(p_pg.head())

# 보기 좋게: 채널별로 따로 그림(그래도 "온/오프라인 비교" 구조 유지)
for ch in df_w["Channel"].dropna().unique():
    cols = [c for c in p_pg.columns if c[0] == ch]
    if len(cols) == 0:
        continue
    tmp = p_pg[cols]
    tmp.columns = [c[1] for c in cols]  # 상품군 이름만
    tmp.rename(index=month_label).plot(kind="bar")
    plt.title(f"겨울 월별×상품군 판매량 - {ch} (Top {TOP_N})")
    plt.xlabel("월")
    plt.ylabel("판매량(건수)")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
