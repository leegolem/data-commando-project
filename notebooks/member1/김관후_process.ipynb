```python
customer_channel_summary = (
    df1.groupby(["customer_id", "channel_label"])
       .agg(
           channel_spend=("price", "sum"),
           channel_purchases=("price", "size")  # 행 개수 = 구매건수
       )
       .reset_index()
)

customer_channel_summary.head()
```

```python
customer_channel_summary = customer_channel_summary.merge(
    customer_summary[["customer_id", "total_spend", "segment"]],
    on="customer_id",
    how="left"
)

customer_channel_summary["channel_share"] = (
    customer_channel_summary["channel_spend"] / customer_channel_summary["total_spend"]
)

customer_channel_summary.head()
```

```python
customer_channel_summary["channel_label"].value_counts(dropna=False)
customer_channel_summary.groupby("customer_id")["channel_label"].nunique().value_counts().head(10)
```

```python
# 1)
wide = (
    customer_channel_summary
    .pivot_table(
        index="customer_id",
        columns="channel_label",
        values=["channel_spend", "channel_purchases"],
        aggfunc="sum",
        fill_value=0
    )
)

# 2) 컬럼 이름 깔끔하게 만들기
wide.columns = [f"{v}_{k}" for v, k in wide.columns]  
wide = wide.reset_index()

wide.head()
```

```python
wide = wide.merge(
    customer_summary[["customer_id","total_spend","segment"]],
    on="customer_id",
    how="left"
)

wide["online_share"]  = wide["channel_spend_온라인"]  / wide["total_spend"]
wide["offline_share"] = wide["channel_spend_오프라인"] / wide["total_spend"]
```
