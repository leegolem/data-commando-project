## 3가지 데이터 join

```python
import pandas as pd

articles = pd.read_csv('./data/cleaned_articles_hm.csv') # Article_Id
customer = pd.read_csv('./data/cleaned_customer_hm.csv') # Customer_Id
transactions = pd.read_csv('./data/cleaned_transactions_hm.csv') # 메인 테이블
```

```python
merged_df = transactions.merge(articles, on='Article_Id', how='inner') \
                        .merge(customer, on='Customer_Id', how='inner')
merged_df.head()
```
