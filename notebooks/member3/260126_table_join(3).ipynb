## 3가지 데이터 join

```python
import pandas as pd

articles = pd.read_csv('./data/cleaned_articles_hm.csv') # Article_Id
customer = pd.read_csv('./data/cleaned_customer_hm.csv') # Customer_Id
transactions = pd.read_csv('./data/cleaned_transactions_hm.csv') # 메인 테이블
```

```python
merged_df = transactions.merge(articles, on='Article_Id', how='inner') \
                        .merge(customer, on='Customer_Id', how='inner')
merged_df.head()
```

### 매출 상위 20%, 하위 20% 태그

```python
customer_revenue = merged_df.groupby('Customer_Id')['Revenue'].sum().reset_index()
```

```python
top = customer_revenue['Revenue'].quantile(0.8) # 매출 80% 지점이 어디인지
bottom = customer_revenue['Revenue'].quantile(0.2) # 매출 20% 지점이 어디인지
```

```python
top_20 = customer_revenue[customer_revenue['Revenue'] >= top]
bottom_20 = customer_revenue[customer_revenue['Revenue'] <= bottom]
```

```python
print(f"전체 고객 수: {len(customer_revenue)}명")
print(f"상위 20% 고객 수: {len(top_20)}명")
print(f"상위 20% 최소 매출액: {top}")

print(f"하위 20% 고객 수: {len(bottom_20)}명")
print(f"하위 20% 최대 매출액: {bottom}")
```

```python
total_revenue = customer_revenue['Revenue'].sum()
top_20_total_revenue = top_20['Revenue'].sum()

ratio = (top_20_total_revenue / total_revenue) * 100
print(f"상위 20% 고객의 매출 비중: {ratio:.2f}%")
```

```python
import numpy as np

# 0~1 사이의 백분위수 계산 (1에 가까울수록 고매출)
customer_revenue['percentile'] = customer_revenue['Revenue'].rank(pct=True)

# 조건 설정
# 상위 20% = percentile 0.8 이상
# 하위 20% = percentile 0.2 이하
conditions = [
    (customer_revenue['percentile'] >= 0.8),
    (customer_revenue['percentile'] <= 0.2)
]
choices = ['Top_20', 'Bottom_20']

# 'Segment' 컬럼 생성 (나머지는 'Middle'로 채움)
customer_revenue['Segment'] = np.select(conditions, choices, default='Middle')
```

```python
customer_revenue
```

```python
merged_df = pd.merge(merged_df, customer_revenue[['Customer_Id', 'Segment']], on='Customer_Id', how='inner')
```

### 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Top 20과 Bottom 20 집단 간 총 매출 비교

```python
segment_revenue = merged_df.groupby('Segment')['Revenue'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=segment_revenue, x='Segment', y='Revenue')
plt.title('Top 20 vs Bottom 20')
plt.ylabel('Total Revenue')
plt.show()
```

#### Top 20, Bottom 20별 구매 채널 비교

```python
segment_channel = merged_df.groupby(['Segment','Sales_Channel_Id']).size().unstack()  # groupby를 하면 두개 컬럼 다 행에 있는데, 피벗테이블처럼 구분하기 위해서 .unstack()을 사용해서 Channel 값을 열로 바꿔줌
segment_channel_pct = segment_channel.div(segment_channel.sum(axis=1), axis=0) * 100  # sum(axis=1)으로 여러개 열을 하나로 합침 → segment별로 sum, axis=0 기준으로 div 해서 channel별로 비율 나눔

segment_channel_pct.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Segment-Channel(%)')
plt.ylabel('Percentage(%)')
plt.legend(title='Channel (1 or 2)')
plt.xticks(rotation=0)
plt.show()
```

#### Top 20, Bottom 20별 Fashion News 구독 여부 비교

```python
segment_FN = merged_df.groupby(['Segment','FN']).size().unstack()  # groupby를 하면 두개 컬럼 다 행에 있는데, 피벗테이블처럼 구분하기 위해서 .unstack()을 사용해서 FN 값을 열로 바꿔줌
segment_FN_pct = segment_FN.div(segment_FN.sum(axis=1), axis=0) * 100  # sum(axis=1)으로 여러개 열을 하나로 합침 → segment별로 sum, axis=0 기준으로 div 해서 FN별로 비율 나눔

segment_FN_pct.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Segment-FN(%)')
plt.ylabel('Percentage(%)')
plt.legend(title='FN (0 or 1)')
plt.xticks(rotation=0)
plt.show()
```
