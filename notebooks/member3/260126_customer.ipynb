### 나이 그룹 분류
10대/20대/30대/40대/50대/60대+(60대~70대)

```python
# 나이를 10단위로 나누어 연령대 컬럼 생성
df_drop['age_group'] = (df_drop['age'] // 10) * 10

# 연령대별 고객 분포 확인
print(df_drop['age_group'].value_counts().sort_index())
```

60대와 70대를 합쳐도 약 7만건 정도라서 10대 제외 나머지 연령대랑 비교하면 숫자가 적은 편임  
**→ 60대 이상으로 한번에 그룹 짓기**

```python
# 60세 이상은 모두 60으로 통일 (60대 이상 그룹화)
df_drop['age_group'] = df_drop['age'].apply(lambda x: 60 if x >= 60 else (x // 10) * 10)

# 결과 확인
print(df_drop['age_group'].value_counts().sort_index())
```

### FN 값과 fashion_news_frequency 간 관계 파악

**FN**  
**0    675069**  
1    372358  
Name: count, dtype: int64  

**fashion_news_frequency**  
**None         674002**  
Regularly    372767  
Monthly         658  
Name: count, dtype: int64  

→ FN에서 0 개수와 fashion_news_frequency에서 None 개수가 맞지 않음

차이가 발생하게 된 이유 추측:  
FN(뉴스 구독)은 0으로 해놓곤, 뉴스 수신 빈도는 Regularly 또는 Monthly로 설정  
**→ 어쨌든 0으로 설정해뒀으면 뉴스 구독 안하는거니까 FN==0 이라면 fashion_news_frequency도 None으로 변경해줘야 함**

```python
new_df.loc[new_df['FN'] == 0, 'fashion_news_frequency'] = 'None'
```

```python
new_df['fashion_news_frequency'].value_counts()
```

```python
new_df['FN'].value_counts()
```

그런데 왜 FN 0인 행 수보다 fashion_news_frequency None인 행 수가 더 많을까?  
→ FN 1인데 주기 설정을 아직 안해서 뉴스 못 받는 경우도 존재

```python
# FN이 1인데도 frequency가 'None'인 사람들의 수
mismatch_count = len(new_df[(new_df['FN'] == 1) & (new_df['fashion_news_frequency'] == 'None')])

print(f"수신동의는 했지만 설정은 None인 사람: {mismatch_count}명")
```

쇼핑 뉴스를 받은 사람과 받지 않은 사람을 확실히 구분하기 위해서  
위 624명에 대해 FN 0으로 수정한다

```python
new_df.loc[(new_df['FN'] == 1) & (new_df['fashion_news_frequency'] == 'None'), 'FN'] = 0
```

```python
new_df['fashion_news_frequency'].value_counts()
```

```python
new_df['FN'].value_counts()
```

**이제 FN이 0이라면 fashion_news_frequency도 None으로 맞춰진 상태**
