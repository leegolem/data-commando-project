## 개별 상품의 카테고리, 색상, 유형 등 상품군의 특성이 기록된 테이블

- 상품 식별자
    - article_id                    # 상품을 식별하는 ID
    - product_code                  # 상품 코드
    - prod_name                     # 상품명

<br>

- 상품 타입/제품군
    - product_type_no	            # 상품 종류 코드
    - product_type_name	            # 상품 종류 명

<br>

- 상품 유형 / 카테고리 (가장 중요)
    - garment_group_no	            # 대분류 코드
    - garment_group_name	        # 대분류 명 - (예시) Socks and Tights
    - index_code                    # 중분류 코드
    - index_name                    # 중분류 명 (예: Lingeries/Tights)
    - index_group_no                # 타겟 그룹 코드
    - index_group_name              # 타겟 그룹 명 (예: Ladieswear)
    - section_no                    # 섹션 코드
    - section_name                  # 섹션 명 (예: Womens Nightwear, Socks & Tigh)
    - department_no                 # 소분류 코드
    - department_name               # 소분류 명 (예: Tights basic)
- 상품군별 매출 분석 가능

<br>

- 색상 / 외형 정보
    - graphical_appearance_no       # 외형 정보 코드 
    - graphical_appearance_name     # 외형 스타일 - (예시) Solid
    - colour_group_code             # 상품 색상 코드
    - colour_group_name             # 상품 색상 명 - (예시) Light Beige
    - perceived_colour_value_id     # 명도 id - 명도 분류 - (예시) Dusty Light
    - perceived_colour_value_name   # 명도
    - perceived_colour_master_id    # 색상군 코드
    - perceived_colour_master_name  # 색상군 - (예시) Beige    
- 색상·타입별 매출 분포는 어떻게 되어 있는가?

<br>

- 텍스트 설명
    - detail_desc                   # 상품 상세 설명
