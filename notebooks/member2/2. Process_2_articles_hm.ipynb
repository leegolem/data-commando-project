# 전처리

```python
art = articles.copy()
```

```python
use_cols = [
    "article_id",                    # 조인 키
    "product_code",                  # (선택) 모델/라인 단위 묶음에 유용
    "product_type_no",
    "product_type_name",             # 상품 유형
    "product_group_name",            # 상품 그룹(상위)
    "garment_group_name",            # 의류 그룹(핵심)
    "index_group_no",                # 타겟 그룹 코드(핵심)
    "index_group_name",              # 타겟 그룹(핵심)
    "index_name",                    # 중분류(선택)
    "section_no",                    # 섹션 코드(추천)
    "section_name",                  # 섹션(추천)
    "graphical_appearance_name",     # 외형 스타일(선택)
    "colour_group_name",             # 색상(선택)
    "perceived_colour_value_name",   # 명도(선택)
    "perceived_colour_master_id",    # 색상군 코드(선택)
    "perceived_colour_master_name",  # 색상군(선택)
    "prod_name"                      # 상품명(설명용, 분석변수로는 비추천)
    
]
art = art[use_cols]
```

code와 name을 합치지 않은 이유
- 범주가 불필요하게 늘어짐

- 코드를 그냥 남기는것도 일종의 방법처럼 인식됨 (단 상위코드만)

추가적으로 남길만한 컬럼


- product_type_no: 상품 분류이기 때문
- index_group_no: 상위 타깃 카테고리이기 때문에 중요 (Ladieswear/Menswear/Kids)과 같은 분류이기 때문
- section_no: 실제 매장/판매 섹션이기 때문에 필요하다 판단

- perceived_colour_master_id: 색상군엔 (Black vs Dark Black) 같이 비슷한 내용이 있을수도 있음



```python
art = art.rename(columns={
    # code / id
    "product_type_no": "product_type_code",
    "index_group_no": "index_group_code",
    "section_no": "section_code",
    "perceived_colour_master_id": "color_master_code",

    # name
    "product_code": "product_code",
    "product_type_name": "product_type",
    "product_group_name": "product_group",
    "garment_group_name": "garment_group",
    "index_group_name": "index_group",
    "index_name": "index_name",
    "section_name": "section",
    "graphical_appearance_name": "graphic",
    "colour_group_name": "color",
    "perceived_colour_value_name": "tone",
    "perceived_colour_master_name": "color_master",
    "prod_name": "prod_name"
})
```

```python
fill_cols = [
    "product_type", "product_group", "garment_group",
    "index_group", "index_name", "section",
    "graphic", "color", "tone", "color_master"
]
art[fill_cols] = art[fill_cols].fillna("Unknown")
```

```python
# code 컬럼: 숫자로 유지하면서 결측만 채움
code_cols = [
    "product_type_code",
    "index_group_code",
    "section_code",
    "color_master_code"
]

for c in code_cols:
    if c in art.columns:
        art[c] = pd.to_numeric(art[c], errors="coerce").fillna(-1).astype("int64")
```
